### Selection Sort 

```python
num = [8,4,2,1,3,5,6,10,9,7]

def func(num) : 
    n = len(num) 
    for i in range(0 , n) :
        min_index = i
        for j in range(i+1 , n):
            if num[j] < num[min_index]:
                min_index = j
        num[i] , num[min_index] = num[min_index], num[i]
    return num
 
func(num=num)
```

## Bubble Sort

```python
array = [8,4,2,1,3,5,6,10,9,7]


def func(arr) :
    n = len(arr)
    is_sort_done = False
    for i in range(n-2 , -1 , -1):
        for j in range(0 , i+1) : 
            if arr[j+1] < arr[j] : 
                is_sort_done = True
                arr[j+1] , arr[j] = arr[j], arr[j+1]
        if is_sort_done == False :
            break
    return arr

func(arr = array)
```
