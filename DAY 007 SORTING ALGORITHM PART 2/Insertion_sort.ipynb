```python
num = [3,5,6,4,8,9,10,7,1]

def func(num) :
    n = len(num)
    for i in range(1 , n):
        j = i-1 
        key = num[i]
        
        while j >=0 and num[j] > key :
            num[j+1] = num[j]
            j = j-1
        num[j+1] = key
    return num
        
func(num = num)
```
