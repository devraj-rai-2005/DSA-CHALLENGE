## Merging the sorted Array's

```python
left = [1,2,4]
right = [8,9,10]

def func(l , r):
    re = []
    n1 = len(l)
    n2 = len(r)
    i , j = 0 ,0 
    while i<n1 and j <n2 :
        if l[i] <= r[j]:
            re.append(l[i])
            i = i+1
        else:
            re.append(r[j])
            j = j+1
            
    if i < n1 :
        while i < n1:
            re.append(l[i])
            i = i+1 
            
    if j <n2 :
        while j <n2:
            re.append(r[j])
            j = j+1
    return re

func(l = left , r = right)
```
