## Merging of the Sorted Array.

```python
# Q1 


arr1 = [1, 3, 5]
arr2 = [2, 4, 6]


def func(l, r) :
    re = []
    m = len(l)
    n = len(r)
    i = 0 
    j = 0
    
    while i < m  and j < n :
        if l[i] <= r[j] :
            re.append(l[i])
            i =i+1
        else:
            re.append(r[j])
            j= j+1
    
    if i < m : 
        while i < m :
            re.append(l[i]) 
            i = i +1 
            
    if j < n : 
        while j <n  :
            re.append(r[j])
            j = j + 1
            
    
    return re


func(l=arr1 , r = arr2)
```
