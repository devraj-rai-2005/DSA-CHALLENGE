#### Q1 :- Move Zero to Last

### My otimal Solution

```python
num =[0,1,0,3,12]


def func(num) :
    i= 0 
    n = len(num)
    re = []
    zero_list = []
    
    while i < n :  
        if num[i] == 0 :
            zero_list.append(num[i])
            i = i+1 
        else :
            re.append(num[i])
            i= i+1 
    num[:] = re + zero_list
        

func(num=num)
num
```

### Teacher Optimal Solution

```python
num =[4,2,4,0,0,3,0,5,1,0]

def func(a1) :
    n= len(a1) 
    i = 0 
    j = i+1 
    
    while i < n and j < n :
        if len(a1) == 1 :
            return
        if a1[i] != 0 :
            i = i+ 1
            j = j+1
        if a1[i] == 0 :
            if a1[j] !=0 :
                a1[i] ,a1[j] =a1[j] , a1[i]
                i = i+1
                j = j+1 
                
            else : 
                j = j+1 
            
    
func(a1 =num)
num
```
