### Linear Search

```python
num_1 = [5,3,5,9,8,1,6,4,-10,-100]

num_2 = [10,20,25 ,10,10,-5,-3,-2,7]

num_3 = [5,3,2,5,6,7,10,1]

def func(array, target) :
    n = len(array)
    
    for i in range(0 , n) :
        if array[i] == target :
            return i 
    return -1

func(array=num_3 , target= 100)
```
