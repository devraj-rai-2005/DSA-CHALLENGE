### Merge Two Array with the removal of the duplicate

```python
array_1 = [1,1,1,2,4,6,7]

array_2 = [1,2,3,6,7,8,9,10, 10,10,11,11,12]


def merge_sort(left , right) : 
    n1 = len(left) 
    n2 = len(right)
    re = []
    i , j = 0,0
    
    while i < n1 and j < n2 :
        if left[i] <= right[j] :
            re.append(left[i])
            i = i+1
        else: 
            re.append(right[j])
            j = j+1
            
    if i<n1 :
        while i < n1 :
            re.append(left[i])
            i = i+1
      
      
    if j < n2 :
        while j  < n2 :
            re.append(right[j])
            j= j + 1
    re
            
    
    m = 0 
    n = m+1
    ar =[]
    n3 = len(re)
    
    while m< n3 and n < n3 :
        if re[m] != re[n] :
            ar.append(re[m])
            m = n 
            n = n+1
        else :
            n = n+1 
        if n == n3 :
            ar.append(re[-1])
    return ar


merge_sort(left=array_1 , right=array_2)
```
