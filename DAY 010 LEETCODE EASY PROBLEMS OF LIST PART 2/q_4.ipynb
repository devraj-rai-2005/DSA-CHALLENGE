## LeetCode

```python
def missing_func(array) :
    n = len(array)
    
    for i in range(0 , n+1) :
        if i not in  array:
            missing = i 
    return missing


ar1 = [1,0,3,4]
ar2 = [9,6,4,2,3,5,7,0,1,3]
ar3 = [3,0,1]

missing_func(array=ar2)
```
