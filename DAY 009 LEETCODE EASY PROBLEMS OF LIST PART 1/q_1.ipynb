## Q1  Find Largest Element in a List / Array

```python
## Method 1 


array = [55,32,-97 , 99 , 3 , 67]

def merge_func(left , right) :
    i, j = 0,0 
    n1 = len(left)
    n2= len(right)
    re = []
    
    while i < n1 and j < n2 :
        if left[i] <= right[j] :
            re.append(left[i])
            i = i+ 1 
        else :
            re.append(right[j])
            j = j + 1 
    if i < n1 :
        while i < n1 :
            re.append(left[i])
            i = i+1 
    if j < n2 :
        while j < n2 :
            re.append(right[j])
            j = j+1 
            
    return re

def sort_func(array) :
    
    if len(array) <= 1 :
        return array
    
    n = len(array)
    mid = n//2
    
    left_array = array[ : mid]
    right_array = array[mid :]
    left = sort_func(array = left_array)
    right = sort_func(array=right_array)
    
    return merge_func(left=left , right=right)


x = sort_func(array=array)

print(x[-2])
```

```python
### Method 2 


num = [55,32, -97 , 99 , 3 , 67]

def func(array):
    largest = float("-inf")
    n = len(array)
    
    for i in range(0 ,n):
        if array[i] > largest:
            largest = array[i]
    return largest
        
        
        
func(array=num)
```
