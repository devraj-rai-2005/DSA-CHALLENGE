### Rotate the List in the right

```python
num = [1,2,3,4,5,6,7,8,9]
n = len(num)
num[:] = [num[-1]] + num[0 : n-1]
```

```python
num = [3,9,5,6,7,2]


def func(array) :
    k = 6
    n = len(array)
    rotation = k%n
    for _ in range(0,rotation):
        if k == n :
            return array
        ld = array[n-1]
        for i in range(n-2 , -1 , -1): 
            array[i+1] = array[i]
            
        array[0] = ld 
    print(array)

func(array=num)
```
