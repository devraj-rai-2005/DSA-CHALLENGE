## Check Whether the sorted array will not carry a duplicate.

```python
num = [1,1,1,2,3,4,4,7,9,9,9,10]

def func(array) : 

    i = 0 
    j  = i +1 

    n = len(array)

    while i < n and j < n :
        if array[i] == array[j] :
            j = j+1 
        else :
            i = i+1 
            array[i] , array[j] = array[j] , array[i]
            j = j+1 
    return  array

func(array=num)
```
