## Number is Palindrome

```python
n = 121
j = n 
de = 0

while n > 0 :
    ld = n%10 
    de = (de * 10 ) + ld
    n = n//10

j == de
```

## String Palindrome

```python
string  = 'abcdeedcbl'

s = string

def func(s , l , r):
    if l>=r : 
        return True
    if s[l] == s[r]:
        return func(s , l = l+1 , r = r-1)
    else : 
        return False
```

```python
func(s = s , l = 0 , r = len(s) - 1)
```
