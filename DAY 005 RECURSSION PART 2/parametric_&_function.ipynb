## Parameterized Recurssion

```python
## Sum of N Natural Numbers 

def func(n , total):
    if n == 11:
        print(total)
        return 
    total = total +n 
    func(n=n+1 , total=total)
```

```python
func(n = 1, total=0)
```

## Functional Recurssion

## Sum of all N natural Numbers 

```python
def func(n):
    if n == 1 :
        return 1 
    return n + func(n = n-1)

x = func(n = 4)

print(x)
```
