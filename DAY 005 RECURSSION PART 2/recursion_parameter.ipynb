### Print the n time x 

```python
def func(x , n):
    if n == 0 :
        return 
    print(x , n)
    func(n = n - 1 , x = x)
func(n = 4 , x = "dev")
```

### Print 1 to n 

##### n = 5

```python
## Tail Recursion

def func(x , n):
    if n ==  6:
        return 
    print(x , n )
    func(x = x , n = n+1)

func(x = "dev" , n = 1)
```

```python
### Head Recursion

def func(x, n):
    if n == 0 :
        return 
    func(x , n= n -1)
    print(x , n)

func(x = "dev" , n = 4)
```

```python
## N -- 1 

#  With the tail recursion


def func(n):
    if n== 0 :
        return 
    print(n)
    func(n - 1) 
    
    
func(n = 4)
```

```python
# Head Recursion

def func(n):
    if n== 0 :
        return 
    func(n = n-1) 
    print(n)
    
    
func(n = 4)
```
