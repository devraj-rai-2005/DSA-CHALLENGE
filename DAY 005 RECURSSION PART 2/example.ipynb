```python
## Q9 Check whether an array is sorted.

array =  [1,2,3,4]

def func(arr , n ):
    if n == len(arr)-1 : 
        return True
    
    if  arr[n] >= arr[n+1]:
        return False
    return func(arr = array , n = n+1)

func(arr=array , n= 0)
```

```python
# Q12  

ar = [2,5,2,7,2]

def func(ar , n , i , t ):
    if i == len(ar) :
        return n
    if ar[i] == t :
        return func(ar = ar , n = n+1 , i= i+1 , t = t)
    else :
        return func(ar = ar , n = n, i= i+1 , t = t)
```
