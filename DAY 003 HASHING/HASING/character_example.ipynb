```python
# Find the frequency of only these characters.

s = "banana"
q = ['a', 'b', 'n', 'z']

# Expected:

{
    'a': 3,
    'b': 1,
    'n': 2,
    'z': 0
}
```

```python
s = "banana"
q = ['a', 'b', 'n', 'z']

freq = {}
ans = {}

for ch in s :
    freq[ch] = freq.get(ch , 0)+1
for c in q:
    ans[c] = freq.get(c , 0)
ans
```

## Q3

```python
# Print only the characters that appear exactly once.

s = "mississippi"

# Expected output:

m
```

```python
s = "mississippi"
freq = {}
ans = {}

for ch in s:
    freq[ch] = freq.get(ch , 0)+1
for i in freq:
    if freq[i] == 1 :
        ans[i] = freq[i]
ans
```

## Q4

```python
# Print only the characters that appear more than once.

s = "programming"

# Expected output (order may vary):

# r
# g
# m
```

```python
s = "programming"
freq = {}
ans = {}

for ch in s:
    freq[ch] = freq.get(ch , 0)+1
for j in freq :
    if freq[j]> 1:
        ans[j] = freq[j]
    
ans
```

## Q10

```python
s = "leetcode"

queries = ['l','e','t','z','o']

# Output:

{
'l':1,
'e':3,
't':1,
'z':0,
'o':1
}
```

```python
s = "leetcode"

queries = ['l','e','t','z','o']

freq = {}

ans = {}

for j in s:
    freq[j] = freq.get(j , 0)+1

for ch in queries:
    ans[ch] = freq.get(ch , 0)
ans
```
