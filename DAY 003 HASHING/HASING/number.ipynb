* Note :- Hashing is for storing the values in the data-structures like the array , dict , sets and then we have to fetch it.

## Brute Approach

```python
a = [5,3,2,2,1,5,5,7,5,10]
b = [10,111,1,9,5,67,2]

d = dict()

for num in b :
    count = 0 
    for n in a:
        if num == n :
            count +=1
    d[num] = count
    
d
```

## Optimal Method

```python
a = [5,3,2,2,1,5,5,7,5,10]
b = [10,111,1,9,5,67,2]
hash_list = [0] *11
de = {}

for num in a:
    hash_list[num]+=1

for n in b :
    if 0<n<11:
        de[n] = hash_list[n]
    else : 
        de[n] = 0
        
print(de)
```

## BETTER APPROACH

```python
a = [5,3,2,2,1,5,5,7,5,10]
b = [10,111,1,9,5,67,2]

hash_map = {}
answer = {}

for num in a:
   hash_map[num] = hash_map.get(num, 0) + 1
   print(hash_map)
   

for n in b :
    answer[n] = hash_map.get(n ,0)
    print(answer)
```
