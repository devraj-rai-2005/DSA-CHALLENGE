## Example 7

```python

a = [2,3,5,5,7]
b = [5,7,9]

# Output

{
5:2,
7:1,
9:0
}
```

```python
a = [2,3,5,5,7]
b = [5,7,9]
de = {}
hash_list = [0]*8

for num in a:
    hash_list[num]+=1
print(hash_list)

for n in b:
    if 0<n<8:
        de[n] = hash_list[n]
    else:
        de[n] = 0
de
```

## Example 8

```python

a = [1,1,2,2,3,4,4]

# Find which numbers appear only once.

# Output

3
```

```python

b = [1,1,2,2,3,4,4]

a = b.copy()

hash_list = [0]*5

for num in a :
    hash_list[num]+=1

for i in range(len(hash_list)):
    if hash_list[i] == 1:
        print(i)
```
