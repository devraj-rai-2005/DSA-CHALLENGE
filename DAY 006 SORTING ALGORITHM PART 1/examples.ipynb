## Selection Sort

```python
# Ascending

num = [5,7,2,1,3,4,8]

arr = num

def func(arr):
    n = len(arr)
    for i in range(0, n) :
        min_index = i
        for j in range(i+1, n) :
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i] , arr[min_index]  = arr[min_index] , arr[i]
    return arr
func(arr)
```

```python
#### Descending


num = [5,7,2,1,3,4,8]

arr = num

def func(arr):
    n = len(arr)
    for i in range(0, n) :
        min_index = i
        for j in range(i+1, n) :
            if arr[j] > arr[min_index]:
                min_index = j
        arr[i] , arr[min_index]  = arr[min_index] , arr[i]
    return arr
func(arr)
```

## Bubble Sort

```python
num = [1, 3, 2, 4, 5, 7, 8]

ar = num

def func(ar) :
    is_sort_done = False
    n = len(ar)
    for i in range(n-2 , -1 ,-1 ) :
        for j in range( 0 , i+1 ):
            if ar[j+1] < ar[j] :
                is_sort_done = True
                ar[j] , ar[j+1] = ar[j+1] , ar[j]
        if is_sort_done == False :
            break
    return ar

func(ar=ar)
```
