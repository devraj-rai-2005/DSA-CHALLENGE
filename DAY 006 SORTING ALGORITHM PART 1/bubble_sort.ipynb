```python
num = [1, 2, 4, 5, 6, 8, 9]

def func(num ):
    n = len(num)
    i = n-2
    for i in range(n-2 ,-1, -1) :
        is_swap = False
        for j in range(0,  i+1) :
            if num[j+1] < num[j] :
                num[j+1] , num[j] = num[j] , num[j+1]
                is_swap =True
        if is_swap == False :
            break 
        
func(num=num)
```
