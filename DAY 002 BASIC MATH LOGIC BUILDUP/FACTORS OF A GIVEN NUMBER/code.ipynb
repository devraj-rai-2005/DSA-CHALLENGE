## BRUTE APPROACH

```python
num = 20
```

```python
n = num 

result = []


for i in range(1 , n+1):
    if n%i == 0 :
        result.append(i)
    result
    
result
```

## BETTER APPROACH

```python
num = 20

n = num 

result = []

for i in range(1 , num//2+1):
    if num%i == 0 :
        result.append(i)
result.append(num)
print(result)
```

## Optimal Solution

```python
from math import sqrt

n = 10 
num = n

result=  []
for i in range(1 , int(sqrt(num))+1):
    if num%i == 0 :
        result.append(i)
        if num//i != i:
            result.append(num//i)
    result.sort()
result
```
