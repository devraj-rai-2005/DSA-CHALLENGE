## Q1 

Print numbers from 1 to 5.

✅ Head Recursion

✅ Tail Recursion

### HEAD RECURSION

```python
def func(count ):
    if count ==0:
        return
    func(count-1)
    print(count)
```

```python
func(count=5)
```

## Tail Recursion

```python
def func(count):
    if count == 6 :
        return 
    print(count)
    func(count + 1)
```

```python
func(count=1)
```
